```python
 data_russia_losses_equipment['mobile SRBM system'].mean()
```

```python
#rellenamos las columnas con el valor de la media
data_russia_losses_equipment['special equipment'] = data_russia_losses_equipment['special equipment'].fillna(data_russia_losses_equipment['special equipment'].mean())
```

```python
data_russia_losses_equipment['mobile SRBM system'] = data_russia_losses_equipment['mobile SRBM system'].fillna(data_russia_losses_equipment['mobile SRBM system'].mean())
```

```python
#verificamos que ya no existen nulos
data_russia_losses_equipment.info()
```

```python
# Vamos a unir las dos dataframe
#datas_russia = pd.concat([data_russia_losses_equipment, data_russia_losses_personnel],axis=1)
datas_russia = data_russia_losses_equipment.merge(data_russia_losses_personnel,on=('date','day') ,how="left")
datas_russia 
```

# **GRAFIQUEMOS**
